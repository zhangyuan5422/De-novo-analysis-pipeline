```python
#read annotation for 1Kb 
anno=pd.read_csv('/home/zhangy15/projects/SSC_SPARK_combine/SSC_SPARK_3_22/anno.hg38_multianno.csv', sep=',')
#add column names
anno=anno[['Chr','Start','End','Func.refGene','Gene.refGene','GeneDetail.refGene']]
```

```python
#intergenic gene
anno_gt=anno[anno['Func.refGene'] =='intergenic']
anno_gt
```

```python
#split gene and dist
splits=anno_gt['Gene.refGene'].str.split(';', expand=True)
anno_gt['geneA'] = splits[0]
anno_gt['geneB'] = splits[1]
#split dist
splitsA=anno_gt['GeneDetail.refGene'].str.split(';', expand=True)
anno_gt['distA'] = splitsA[0]
anno_gt['distB'] = splitsA[1]
#replace 'dist=' with ''
anno_gt['distA']=anno_gt['distA'].str.replace('dist=', '')
anno_gt['distB']=anno_gt['distB'].str.replace('dist=', '')
# Ensure distA and distB are numeric, handling 'NONE' values
anno_gt['distA'] = pd.to_numeric(anno_gt['distA'], errors='coerce')
anno_gt['distB'] = pd.to_numeric(anno_gt['distB'], errors='coerce')

def decide_gene(row):
    if pd.isnull(row['distA']) and not pd.isnull(row['distB']):
        return row['geneB']
    elif not pd.isnull(row['distA']) and pd.isnull(row['distB']):
        return row['geneA']
    elif not pd.isnull(row['distA']) and not pd.isnull(row['distB']):
        if row['distA'] > row['distB']:
            return row['geneB']
        else:
            return row['geneA']
    else:
        return np.nan  # Or another default value for cases where both are NONE

# Apply the function to decide the gene based on the conditions
anno_gt['gene'] = anno_gt.apply(decide_gene, axis=1)
```

```python
anno_df=anno_gt[['Chr','Start','End','Func.refGene','gene']]
anno_df
```

```python
#non-intergenic gene 
anno_noncoding=anno[anno['Func.refGene'] !='intergenic']
anno_noncoding=anno_noncoding[['Chr','Start','End','Func.refGene','Gene.refGene']]
anno_noncoding.columns=['Chr','Start','End','Func.refGene','gene']
#concat non-intergenic and intergenic
anno_final=pd.concat([anno_noncoding, anno_df])
anno_final=anno_final.sort_values(['Chr','Start'])
#generate element_id to merge
anno_final['element_id']=anno_final['Chr']+'-'+anno_final['Start'].astype(str)+'-'+anno_final['End'].astype(str)
#extract id and gene to merge to 1Kb
anno_gene=anno_final[['element_id','gene','Func.refGene']]
anno_gene
```

```python
import pandas as pd
DNM_gt=pd.read_csv('dnm_1kb_case.csv', sep='\t')
DNM_gt
```

```python
#spark
# calculated fisher exact p value
from numpy import inf
#DNM_gt = DNM_gt.replace(np.inf, 0)
DNM_gt = DNM_gt.replace(np.NAN, 0)

#intergenic cadd >=15
DNM_gt['obs_spark_case_other']=3508-DNM_gt['obs_spark_case']
DNM_gt['obs_spark_ctrl_other']=2188-DNM_gt['obs_spark_ctrl']
##caculated fisherexact p value
import pandas as pd
from scipy.stats import fisher_exact

# Assuming DNM is your existing DataFrame and it's already loaded

# Add new columns for P-value and odds ratio initialized with NaNs
DNM_gt['pvalue_spark'] = pd.NA
DNM_gt['OR_spark'] = pd.NA

for index, row in DNM_gt.iterrows():
    # Construct the contingency table with the current row's data
    table = [[row['obs_spark_case'], row['obs_spark_ctrl']],
             [row['obs_spark_case_other'], row['obs_spark_ctrl_other']]]
    
    # Perform Fisher's Exact Test
    odds_ratio, p_value = fisher_exact(table, alternative='greater')
    
    # Add the results directly to the same row in DNM
    DNM_gt.at[index, 'pvalue_spark'] = p_value
    DNM_gt.at[index, 'OR_spark'] = odds_ratio

# At this point, DNM has two new columns: 'P-value' and 'odds_ratio'

DNM_gt.fillna(np.inf, inplace=True) #if not the excel will wrong
DNM_gt
```
